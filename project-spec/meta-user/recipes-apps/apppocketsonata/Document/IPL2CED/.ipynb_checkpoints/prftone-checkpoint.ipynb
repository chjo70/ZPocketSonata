```python
def F(n, X):
    """
                   1
        F (t) = ------- sin ( (2n+1) t ) 
         n      2n + 1
    """
    return np.sin((2*n+1)*X)/(2*n+1)


def RectWave(n, X):
    """
          n
        Sigma  F (t)
         k=1
    """
    y = np.zeros(X.shape)
    for k in range(n+1):
        y = F(k, X) + y
    return y


X = np.linspace(-5, 100, 1000)
#for i in (1, 5, 50, 100):
plt.plot(X, RectWave(5, X))
plt.show()
```

```python
# 구형파 발생
def Pulses( data, x ):
    toa = 0
    heaviside_ = 0
    for i in data:
        #toa = i[0] + toa
        #print( i[0], i[1] )
        heaviside_ = ( ( np.heaviside(x-i[0], 0 ) - np.heaviside(x-(i[0]+i[1]), 0) ) * ( i[2]) ) + heaviside_
        #print( '-->', heaviside_ )
        
    return heaviside_ + minPA

# PRF 값 추출
def GetPRF( dtoa ):
    if dtoa < MIN_DTOA:
        x = MAX_PRF
    elif dtoa > MAX_DTOA:
        x = MIN_PRF
    else:
        x = ( MAX_PRF - MIN_PRF ) / ( MIN_DTOA - MAX_DTOA )
        x = x * ( dtoa - MIN_DTOA ) + MAX_PRF    
    return x

# 사인파 발생
def SineWave( x ):
    global prf_dtoa
    global prf_data
    global prf_tone_iter
    
    y = 0

    # 데이터 초기화
    try:
        if prf_toa < 0:
            prf_data = next( prf_tone_iter )
            prf_toa = prf_data[0]
        else:
            if x <= prf_toa:
                pass            
            else:
                prf_data = next( prf_tone_iter )
                prf_toa = prf_data[0]
                
        f = GetPRF( prf_data[0] )
        y = np.sin( f * 2 * np.pi * x ) * prf_data[2]    
                
    except StopIteration as e:
        f = GetPRF( prf_data[0] )
        y = np.sin( f * 2 * np.pi * x )    
        # print( 'exception' )
            
    #print( x, f, y )
       
    return y

def MaxToa( data ):

    # prf_tone[-1][0]
    return prf_tone[-1][0]
        
```
